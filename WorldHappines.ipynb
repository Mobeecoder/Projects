```python
happy_df=df.rename(columns = {'Happiness Rank':'happiness_rank', 'Happiness Score':'happiness_score', 'Standard Error':'standard_error', 'Economy (GDP per Capita)':'Economy', 'Health (Life Expectancy)':'Health','Trust (Government Corruption)':'Trust','Dystopia Residual':'dystopia_residual'})
happy_df.head()
```

```python
happy_df.sort_values(by="happiness_rank", ascending=True).head(10)
```

```python
happy_df.sort_values(by="happiness_rank", ascending=False).head(10)
```

```python
plt.rcParams['figure.figsize'] = (15,7)
plt.title('B/W Happiness Score and Economy')
sns.scatterplot(x = happy_df.happiness_score, y = happy_df.Economy, hue = happy_df.Region, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('Economy')
```

```python
economy_region = happy_df.groupby('Region')['Economy'].sum()
economy_region
```

```python
economy_region.plot.pie(autopct = '%1.1f%%')
plt.title('Economy by Region')
plt.ylabel('')
```

```python
#Total countries
total_country = happy_df.groupby('Region')[['Country']].count()
print(total_country)
```

```python
#Dystopia Residual in Regions

dystopia = happy_df.groupby('Region')[['dystopia_residual']].mean()
dystopia
```

```python
# Check for correlation
corrmat = happy_df.corr()
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Dystopia Residual in various Regions')
plt.xlabel('Regions', fontsize =17)
plt.ylabel('Dystopia Residual', fontsize = 17)
plt.xticks(rotation = 35, ha='right')
plt.bar(dystopia.index, dystopia.dystopia_residual)
```

```python
top_10 = happy_df.sort_values(by="happiness_rank", ascending=True).head(10)
bottom_10 = happy_df.sort_values(by="happiness_rank", ascending=False).head(10)
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.Country
axes[0].set_title('Top 10 happiest countries Health')
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x = top_10.Country, y = top_10.Health, ax = axes[0])
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Health')

xlabels = top_10.Country
axes[1].set_title('Top 10 least happiest countries Health')
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x = bottom_10.Country, y = bottom_10.Health, ax = axes[1])
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Health')
```

```python
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x = happy_df.Freedom, y = happy_df.happiness_score, hue = happy_df.Region, s = 200)
plt.legend(loc = 'upper left', fontsize ='12')
plt.xlabel('Freedom')
plt.ylabel('Happiness Score')
```

```python
country = happy_df.sort_values(by='dystopia_residual').head(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Countries with most Dystopia Residuals')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Dystopia Index', fontsize=13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.Country, country.dystopia_residual)
```

```python
country = happy_df.sort_values(by='dystopia_residual').tail(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Countries with the least Dystopia Residuals')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Dystopia Index', fontsize=13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.Country, country.dystopia_residual)
```

```python
#Dystopia vs Happiness
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x = happy_df.happiness_score, y = happy_df.dystopia_residual, hue = happy_df.Region, s = 200)
plt.legend(loc = 'lower left', fontsize = '14')
plt.xlabel('Happiness Score')
plt.ylabel('Dystopia Residuals')
```
